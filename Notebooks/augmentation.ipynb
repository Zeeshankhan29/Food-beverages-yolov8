# Create synthetic data by Data Augmentation for all classes (brightness and contrast)

```python
import os
from PIL import Image
import numpy as np
import cv2
import albumentations as A
import cv2
import os
from datetime import datetime

valid_class = [
 'Apple Royal Gala',
 'Apple Shimla',
 'Ash Gourd Cut',
 'Avocado',
 'Banana Nendran',
 'Banana Poovan',
 'Banana Red',
 'Brinjal Black Big',
 'Brinjal Purple Striped',
 'Broccoli',
 'Cabbage',
 'Cabbage Red',
 'Capsicum',
 'Capsicum Red',
 'Capsicum Yellow',
 'Cauliflower',
 'Cucumber',
 'Dragon Fruits Red',
 'Golden Kiwi',
 'Jackfruit raw',
 'Kiwi Green',
 'Lemon',
 'Mango Neelam',
 'Mango Totapuri',
 'Mangusteens Indian',
 'Mosambi',
 'Musk Melon Yellow',
 'Muskmelon Rock',
 'Orange Valencia',
 'Papaya',
 'Pineapple',
 'Pomegranate Kabul',
 'Pumpkin cut Wrapped',
 'Radish Red',
 'Sapota Round',
 'Snake Gourd',
 'Tomato Apple',
 'Tomato Nadu',
 'Turnip',
 'Watermelon Kiran']


bright = 'bright_img'
for folder_name in os.listdir(os.getcwd()):
    if not folder_name.startswith('.'):
        if folder_name in valid_class:
            jpg_file = os.path.join(os.getcwd(), folder_name,'BrightImages')
            os.makedirs(jpg_file,exist_ok=True)
            
            seg_file = os.path.join(os.getcwd(), folder_name,'BrightSegment')
            os.makedirs(seg_file,exist_ok=True)
            
            jpg_path = os.path.join(os.getcwd(), folder_name, 'JPEGImages')
            seg_path = os.path.join(os.getcwd(), folder_name, 'SegmentationClass')
   
            for count,file in enumerate(os.listdir(jpg_path)):
                file_path = os.path.join(jpg_path,file)
                if os.path.isfile(file_path):

                    if not file.endswith('.png'):
                        continue
                
                    pillow_image = Image.open(file_path)

                    image = np.array(pillow_image)


                    #Crop an image with width and heigh
                    transform = A.Compose([
                    #     A.RandomCrop(width=640, height=480),
                #         A.HorizontalFlip(p=0.5),
                        A.RandomBrightnessContrast(brightness_limit=(0.1,0.2),
                    contrast_limit=(0.05,0.1), p=1.0),
                    ])
                    transformed_image_1 = transform(image=image)['image']

                    file_name, file_extension = os.path.splitext(file)
                    img_path = os.path.join(jpg_file,f"{file_name}_{bright}_{count}.png")
                    
                    transformed_image_1_bgr = cv2.cvtColor(transformed_image_1, cv2.COLOR_RGB2BGR)

                    cv2.imwrite(img_path,transformed_image_1_bgr)
                    
                    # plt.imshow(transformed_image_1)
            else:
                print(f'Image Augmentation completed for {folder_name}')
                    
            for count1,file1 in enumerate(os.listdir(seg_path)):
                file_path1 = os.path.join(seg_path,file1)
                if os.path.isfile(file_path1):

                    if not file.endswith('.png'):
                        continue
                    pillow_image1 = Image.open(file_path1)

                    image1 = np.array(pillow_image1)

                    file_name1, file_extension = os.path.splitext(file1)
                    img_path1 = os.path.join(seg_file,f"{file_name1}_{bright}_{count1}.png")

                    cv2.imwrite(img_path1,image1)
                    
                    # plt.imshow(transformed_image_1)

            else:
                print(f'Segment Augmentation completed for {folder_name}')

else:
    print('*****************************************')
    print('Augmentation completed for all the classes')
```

```python
path =r'C:\Users\Zeeshan.khan\vscode_projects\training'
import os
import shutil
```

```python
for file in os.listdir(path):
    if not file.startswith('.git') and not  file.startswith('augmentation.ipynb'):
        file_path = os.path.join(path,file)
        for folder in os.listdir(file_path):
#             print(folder)
            if folder.startswith('ZoomImages') or folder.startswith('ZoomSegment'):
                try:
                    not_valid = os.path.join(file_path,folder)
                    os.remove(not_valid)
                except:
                    shutil.rmtree(not_valid)
                    print(f'removed directories {not_valid}')
```
