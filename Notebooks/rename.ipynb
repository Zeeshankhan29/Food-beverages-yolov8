```python
jpg_path=[]
seg_path=[]

for file in os.listdir(os.getcwd()):
    if file.startswith('JPEGImages'):
        jpg_path.append(os.path.join(os.getcwd(),file))
        
    if file.endswith('SegmentationClass'):
        seg_path.append(os.path.join(os.getcwd(),file))
```

```python
import os


for file, file1 in zip(os.listdir(jpg_path[0]), os.listdir(seg_path[0])):
    new_file = file.replace(" Apple Granny smith", "Apple Granny smith")
    new_file1 = file1.replace(" Apple Granny smith", "Apple Granny smith")
    
    os.rename(os.path.join(jpg_path[0], file), os.path.join(jpg_path[0], new_file))
    os.rename(os.path.join(seg_path[0], file1), os.path.join(seg_path[0], new_file1))
    
    print(f"Renamed: {file} to {new_file}")
    print(f"Renamed: {file1} to {new_file1}")
```
